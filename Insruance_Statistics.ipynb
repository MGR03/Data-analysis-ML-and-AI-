```python
#4a Do charges of people who smoke differ significantly from the people who don't? - Yes charges of people who are non smokers is lower than smokers.

sns.violinplot(x='smoker',y='charges',data=mydata[['smoker', 'charges']]);
```

```python
SmokerMedian= mydata[mydata.smoker == 'yes'][['charges']].median()
NonsmokerMedian= mydata[mydata.smoker == 'no'][['charges']].median()
print('Median of Smoker', SmokerMedian)
print('Median of Non-Smoker', NonsmokerMedian)
```

```python
SmokerMedianM= mydata[mydata.smoker == 'yes'][['charges']].mean()
NonsmokerMedianM= mydata[mydata.smoker == 'no'][['charges']].mean()
print('Mean of Smoker', SmokerMedianM)
print('Mean of Non-Smoker', NonsmokerMedianM)
```

```python
#4b Does bmi of males differ significantly from that of females? - Not significantly.

sns.violinplot(x='sex',y='bmi',data=mydata[['sex', 'bmi']]);
```

```python
BMImale=mydata[mydata.sex == 'male'][['bmi']].median()
BMIfemale= mydata[mydata.sex == 'female'][['bmi']].median()
print('Median of Male', BMImale)
print ('Median of Female', BMIfemale)
```

```python
BMImaleM=mydata[mydata.sex == 'male'][['bmi']].mean()
BMIfemaleM= mydata[mydata.sex == 'female'][['bmi']].mean()
print('Mean of Male', BMImaleM)
print ('Mean of Female', BMIfemaleM)
```

```python
#4c Is the proportion of smokers significantly different in different genders? - It is different, but not significantly.

sns.countplot(data=mydata, x='smoker', hue='sex');
```

```python
Smokers = mydata[mydata.smoker == 'yes']
Only_Smoker = Smokers[['smoker','sex']]
Only_Smoker
```

```python
FemaleSmoker = Only_Smoker[Only_Smoker.sex == 'female']
MaleSmoker =Only_Smoker[Only_Smoker.sex == 'male']
FemaleSmoker, MaleSmoker
```

```python
Total_Smoker = mydata[mydata.smoker == 'yes']["smoker"].count()
Total_Smoker
```

```python
RatioFemale = FemaleSmoker["sex"].count()/Total_Smoker
RatioMale = MaleSmoker["sex"].count()/Total_Smoker
RatioFemale, RatioMale
```

```python
# 4d. Distribution of bmi across women with no children, one child and two children, the same

sns.catplot (data=mydata, y='bmi', x='sex', hue='children', dodge=True, kind='strip');
```

```python
#Keeping only woman with kids

newdata = mydata.drop(mydata["sex"].loc[mydata["sex"]=='male'].index);

newdata
```

```python
newdata3 = newdata.drop(newdata["children"].loc[newdata["children"]==3].index);

newdata3
```

```python
newdata4 = newdata3.drop(newdata3["children"].loc[newdata3["children"]==4].index);

newdata4
```

```python
newdata5 = newdata4.drop(newdata4["children"].loc[newdata4["children"]==5].index);

newdata5
```

```python
## Distribution of bmi across women with no children, one child and two children

sns.catplot (data=newdata5, y='bmi', x='children', kind='strip');

plt.title('Ditribution of BMI across women with no children, one child and two children');
plt.xlabel('Number of Children');
plt.ylabel('BMI');
```

```python
sns.swarmplot(data=newdata5, x='children', y='bmi');
plt.title('Ditribution of BMI across women with no children, one child and two children');
plt.xlabel('Number of Children');
plt.ylabel('BMI');
```
